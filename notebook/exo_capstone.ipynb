```python
# Now attempting to calculate the habitable zone for each star
```

```python
def habitable_zone(star_luminosity):
    inner_bound = 0.75 * (star_luminosity ** 0.5)
    outer_bound = 1.75 * (star_luminosity ** 0.5)
    return inner_bound, outer_bound

# this is assuming star_luminosity is related to star_mass... I will double check this if I use this.
eu['star_luminosity'] = eu['star_mass'] ** 3.5
eu['hz_inner'], eu['hz_outer'] = zip(*eu['star_luminosity'].apply(habitable_zone))

# exoplanets in the habitable zone
eu['in_habitable_zone'] = (eu['semi_major_axis'] >= eu['hz_inner']) & (eu['semi_major_axis'] <= eu['hz_outer'])
habitable_exoplanets = eu[eu['in_habitable_zone']]
print(habitable_exoplanets[['planet_name', 'semi_major_axis', 'star_type']])
```

```python
sol['has_global_magnetic_field'] = sol['has_global_magnetic_field'].map({'Yes': True, 'No': False}).fillna(False)
sol['has_liquid_water'] = sol['has_liquid_water'].map({'Yes': True, 'No': False}).fillna(False)
```

```python
sol.columns
```

```python
sol.dtypes
```

```python
def normalize(series):
    return (series - series.min()) / (series.max() - series.min())

sol['norm_distance_from_sun'] = normalize(sol['distance_from_sun'])
sol['norm_mass'] = normalize(sol['mass'])
sol['norm_gravity'] = normalize(sol['gravity'])
```

```python
sol['magnetic_field_score'] = sol['has_global_magnetic_field'].astype(int)
sol['water_score'] = sol['has_liquid_water'].astype(int)
```

```python
weights = {
    'norm_distance_from_sun': 0.3,
    'norm_mass': 0.2,
    'norm_gravity': 0.2,
    'magnetic_field_score': 0.15,
    'water_score': 0.15}

sol['habitability_score'] = (sol['norm_distance_from_sun'] * weights['norm_distance_from_sun'] +
                            sol['norm_mass'] * weights['norm_mass'] +
                            sol['norm_gravity'] * weights['norm_gravity'] +
                            sol['magnetic_field_score'] * weights['magnetic_field_score'] +
                            sol['water_score'] * weights['water_score'])
```

```python
its_a_habbit = sol.sort_values(by='habitability_score', ascending=False)

print(its_a_habbit[['planet', 'habitability_score']])
```
